```python
def celcius_ke_fahrenheit(celsius):
    fahrenheit = (celsius*9/5) + 32
    return fahrenheit

print(celcius_ke_fahrenheit(0))
print(celcius_ke_fahrenheit (100))
```

```python
def is_genap(bilangan):
    return bilangan % 2 == 0

print(is_genap(4))
print(is_genap(7))
```

```python
def nilai(nilai):
    if nilai >= 70:
        return "lulus"
    else:
        return "gagal"

print(nilai(80))
print(nilai(60))
```

```python
def bilangan(n):
    ganjil = [i for i in range(1, n) if i % 2 != 0]
    return ganjil

print(bilangan(20))
```

```python
def bil_ganjil(angka):
    for i in range(1,angka,2):
     print(i, end=" ")
bil_ganjil(20)
```
